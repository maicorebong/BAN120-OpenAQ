## Clustering for CA dataset for years 2020, 2021, & 2022

```python
# Filter to only years 2020, 2021, and 2022
df_cluster_test = df[df['Year'] >= 2020] 
df_cluster_test_ca = df_cluster_test[df_cluster_test['Country'] == 'CA']

# Keep a copy of data
df_cluster_test_ca_copy = df_cluster_test_ca.copy()
```

```python
# Data Preprocessing

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df_cluster_test_ca = df_cluster_test_ca.apply(le.fit_transform)
df_cluster_test_ca.head()
```

```python
from kmodes.kmodes import KModes
```

```python
km_cao = KModes(n_clusters=3, init = "Cao", n_init = 10, verbose=1)
fitClusters_cao = km_cao.fit_predict(df_cluster_test_ca)
```

```python
# Predicted Clusters
fitClusters_cao
```

```python
clusterCentroidsDf = pd.DataFrame(km_cao.cluster_centroids_)
clusterCentroidsDf.columns = df_cluster_test_ca.columns
```

```python
# Mode of the clusters
clusterCentroidsDf
```

```python
kmodes = km_cao.cluster_centroids_
shape = kmodes.shape
shape
```

```python
cost = []
for num_clusters in list(range(1,5)):
    kmode = KModes(n_clusters=num_clusters, init = "cao", n_init = 10, verbose=1)
    kmode.fit_predict(df_cluster_test_ca)
    cost.append(kmode.cost_)
```

```python
y = np.array([i for i in range(1,5,1)])
plt.plot(y,cost)
```

```python
df_cluster_test_ca = df_cluster_test_ca_copy.reset_index()
```

```python
clustersDf = pd.DataFrame(fitClusters_cao)
clustersDf.columns = ['cluster_predicted']
combinedDf = pd.concat([df_cluster_test_ca, clustersDf], axis = 1).reset_index()
combinedDf = combinedDf.drop(['index'], axis = 1)
```

```python
combinedDf.head()
```

```python
label = km_cao.labels_
combinedDf['cluster_name'] = ['cluster1' if x == 0 else\
                           'cluster2' if x == 1 else\
                           'cluster3' if x == 2 else\
                           'cluster4' for x in label]
```
